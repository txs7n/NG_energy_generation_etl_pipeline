```python
# creating a function to extract data from the TCN website 

def fetch_data(date):
    session = requests.Session()
    session.headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
    }
    retries = 3  
    backoff_factor = 2  
    
    for attempt in range(retries):
        try:
            response = session.get(url, timeout=10)
            soup = BeautifulSoup(response.text, 'html.parser')
            viewstate = soup.find('input', {'id': '__VIEWSTATE'})['value']
            eventvalidation = soup.find('input', {'id': '__EVENTVALIDATION'})['value']
            data = {
                '__VIEWSTATE': viewstate,
                '__EVENTVALIDATION': eventvalidation,
                '__EVENTTARGET': '',
                '__EVENTARGUMENT': '',
                'ctl00$MainContent$txtReadingDate': date.strftime('%Y/%m/%d'),
                'ctl00$MainContent$btnGetReadings': 'Get Generation'
            }
            response = session.post(url, data=data, timeout=10)
            soup = BeautifulSoup(response.text, 'html.parser')
            table = soup.find('table', {'id': 'MainContent_gvGeneration'})
            all_rows = []
            for row in table.find_all('tr')[1:]:
                columns = row.find_all('td')
                row_data = [col.text.strip() for col in columns]
                all_rows.append(row_data)
            df_fetch = pd.DataFrame(all_rows, columns=[th.text for th in table.find('tr').find_all('th')])        
            return df_fetch
        except requests.exceptions.RequestException as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            time.sleep(backoff_factor ** attempt)  
    
    return pd.DataFrame()  
```

```python
# creating a function to clean and transform the data collected

def clean_data(df_fetch):
    if not df_fetch.empty:
        if '#' in df_fetch.columns:
            df_fetch.drop(['#'], axis=1, inplace=True)
        if df_fetch.at[df_fetch.index[-1], 'Genco'] == '':
            df_fetch.drop(df_fetch.index[-1], inplace=True)
        for i in df_fetch.columns:
            if i != 'Genco':
                df_fetch[i] = df_fetch[i].str.replace(',', '').astype(float)
        return df_fetch
    return pd.DataFrame()
```

```python
# suppling arguments for the functions

url = "https://www.niggrid.org/GenerationProfile2"
start_date = datetime.strptime('2021/01/01', '%Y/%m/%d')
end_date = datetime.strptime('2024/05/10', '%Y/%m/%d')
current_date = start_date
```

```python
# a while function to loop through the date period, run all the etl functions and load in csv format 

while current_date <= end_date:
    df_fetch = fetch_data(current_date)
    if not df_fetch.empty:
        df_clean = clean_data(df_fetch)
        df_clean.to_csv(f'energy_gen_data_{current_date.strftime("%Y_%m_%d")}.csv', index=False)
    current_date += timedelta(days=1)
```
